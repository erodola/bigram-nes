```python
import matplotlib.pyplot as plt
from bigram import train, generate_quantized, quantize_transition_matrix
```

```python
dataset = '../data/human_names.txt'
T = train(dataset)
```

```python
T_quantized = quantize_transition_matrix(T)

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(T, cmap='Blues')
plt.title('Original Transition Matrix')
plt.colorbar()
plt.subplot(122)
plt.imshow(T_quantized, cmap='Blues')
plt.title('Quantized Transition Matrix')
plt.colorbar()
plt.show()
```

```python
row_sums = T_quantized.sum(axis=1)
assert all(row_sums == 255), "Quantized rows do not sum to 255!"
```

```python
for _ in range(10):
    name = generate_quantized(T_quantized)
    print(name)
```
